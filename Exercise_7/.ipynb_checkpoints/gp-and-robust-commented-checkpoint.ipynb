# Task 1 Gaussian process regression

```python
import numpy as np
import scipy.sparse.linalg as spl
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm
import time
import math
```

<div style="color: green; font-weight:
bold">The loading and displaing of the initial image is identical</div>

```python
# Load the image
image = Image.open('./cc_90.png')
image = image.convert('L')  # convert image to grayscale
image = np.array(image)

plt.imshow(image, cmap='gray')
plt.show()
```

<div style="color: green; font-weight:
bold">The kernel function is identical, however we used the numpy implementation, whereas the sample solution did it manually. <br>
The matrix A was computed completely differently though. We used an approach that calculated for every train pixel the similarity to each other pixel, whereas the sample soution tuned this approach by only calculating the diagonal values of a and then adding the transpose of a again to it. The latter one results in a much better performance. Apart from that the overall different results shouldn't affect the solution that much. <br>
For computing b, we chunked the function because we otherwise ran out of memory. This is due to the fact that we calculated b for every pixel in the image times every trainng image instead of calculating it for every test pixel times every trainng pixel. This affects the solution and also the performance as it takes much longer and can lead to memory issues, which we eventually faced. <br>
Also we only implemented the gaussian process regression and not the nadaraya-watson regression, which is missing. The gpr is implemented correctly though.</div>

```python

def get_x_and_y(image):
    height, width = image.shape
    y, x = np.nonzero(image)
    pixels = np.column_stack((y, x))
    grayvalues = image[y, x]
    return pixels, grayvalues

def kernel(x, x_prime, gamma, h):
    r = np.linalg.norm(x[:, None] - x_prime, axis=-1) / h**2
    return np.exp(-r ** gamma)



pixels, grayvalues = get_x_and_y(image)

def evaluate(gamma, h , sigma_sq):
    start = time.time()
    # Compute A
    A = kernel(pixels, pixels, gamma, h)
    print("computed a")
    # Compute z
    z = spl.spsolve(A + sigma_sq * np.eye(len(pixels)), grayvalues)
    print("computed z")
    # Fill in missing pixel values
    height, width = image.shape
    y, x = np.mgrid[:height, :width]
    all_pixels = np.column_stack((y.ravel(), x.ravel()))

    chunk_size = 10000  # Adjust as needed
    predictions = np.zeros_like(all_pixels[:, 0], dtype=np.float64)
    num_chunks = len(all_pixels) // chunk_size

    with tqdm(total=num_chunks, desc='Processing chunks') as pbar:
        for i in range(0, len(all_pixels), chunk_size):
            chunk = all_pixels[i:i+chunk_size]
            b = kernel(chunk, pixels, gamma, h)
            predictions[i:i+chunk_size] = b.dot(z)
            pbar.update(1)

    image_filled = predictions.reshape((height, width))
    print(f"Took {time.time()-start} seconds for parameters gamma={gamma}, h={h}, sigma_sq={sigma_sq}")

    plt.imshow(image_filled, cmap='gray')
    plt.show()
    
```

### Below you will find the applied algorithm with different hyper parameters.
it appears that the combination of lower hyperparameters (gamma = 0.1, h=0.2, sigma_sq = 0.5) works best in terms of quality. The training time differs vastly between the different adjustments with no real pattern to be recognizable. 

<div style="color: green; font-weight:

ALso a hyperparameter testing was introduced, however the values for the hyperparameters werent changed that much, only marginally, which lead to results that are not as goood as in the sample solution. However a great improvement can be seen anyway.</div>

## 2.1 RANSAC

<div style="color: green; font-weight: 
bold">Same calculation for r, however sample solution is 
    more elegant.
    Points are not passed to variables in sample solution.
    calculation for center seems to be the same.
    Distance calculation is the same, however we solve it using 
    numpy linalg norm. 

</div>

```python
def circumcircle(points):
    """Use this function to calculate the circumscribed circle or circumcircle of a triangle which is a circle that passes through all three vertices."""

    x1 = points[0][0]
    y1 = points[0][1]
    x2 = points[1][0]
    y2 = points[1][1]
    x3 = points[2][0]
    y3 = points[2][1]

     # Calculate the coordinates of the circumcenter
    d = 2 * (x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2))
    x_circle = ((x1**2 + y1**2) * (y2 - y3) + (x2**2 + y2**2) * (y3 - y1) + (x3**2 + y3**2) * (y1 - y2)) / d
    y_circle = ((x1**2 + y1**2) * (x3 - x2) + (x2**2 + y2**2) * (x1 - x3) + (x3**2 + y3**2) * (x2 - x1)) / d

    # Calculate the radius of the circumcircle
    radius = math.sqrt((x_circle - x1)**2 + (y_circle - y1)**2)

    return np.array([x_circle, y_circle]), radius


def euclidean_distance(x, y): 
    """Euclidean distance between two vectors

    :param x: numpy vector
    :param y: numpy vector
    :return: scalar distance
    """
    return np.linalg.norm(x-y)
```

<div style="color: green; font-weight: 
bold">random selection is basically identical. 
    We also check if random ints are equal since we didn't use unique parameter.  
    The ransac calculation is basically identical.

</div>

```python
def find_circles(data, T=50, epsilon=0.001): 
    best_circle_center_x_y = None
    best_circle_radius = None
    best_inlier_count = 0 # higher is better
    best_circle_inliers_indexes = np.array([None,None])

    for iteration in range(T): 
        # generate three random integers between 0 and length of data
        rand_ints = np.random.randint(low=0, high=data.shape[0], size=(3,))

        # select three points based on random integers
        three_random_points = data[rand_ints]

        # check if all points are unique
        if len(np.unique(three_random_points, axis=0)) == 3:
            # calculate circumcircle
            circle_center_x_y, circle_radius = circumcircle(three_random_points)

            inlier_points_indexes = []
            # claculate distance to circle for each point
            for index, point in enumerate(data): 
                distance_to_circle = euclidean_distance(point, circle_center_x_y)

                # check if point lies inside cicrle + some epsilon fault talerance
                if distance_to_circle < (circle_radius + epsilon):
                    inlier_points_indexes.append(index)

            inlier_count = len(inlier_points_indexes)
            # update circle center and radius if new best circle is found
            if inlier_count > best_inlier_count:
                best_circle_center_x_y = circle_center_x_y
                best_circle_radius = circle_radius
                best_inlier_count = inlier_count
                best_circle_inliers_indexes = inlier_points_indexes

    return best_circle_center_x_y, best_circle_radius, best_circle_inliers_indexes
```

<div style="color: green; font-weight: 
bold">We estimated T way too low, which could explain the bad results for found circles. 
    Also we performed the experiments in a visual way. 
    Our selection of epsilon is also very different to the sample solution. Only 0.01 and 0.001 are identical.

</div>

```python
# Visual experiments by using different epsilons and looking at the results manually
epsilons = [1.5, 1, 0.1, 0.05, 0.02, 0.01, 0.001]
for epsilon in epsilons: 

    ############# calculations #############
    data_points = data
    found_circle_centers = []
    found_circle_radii = []

    for i in range(20):
        # calc circle and indexes of inliers for current circle 
        circle_center, circle_radius, circle_inliers_indexes = find_circles(data_points, epsilon=epsilon)
        found_circle_centers.append(circle_center)
        found_circle_radii.append(circle_radius)

        # delete all inliers of the last fitted circle
        data_points = np.delete(data_points, circle_inliers_indexes, axis=0)

        if data_points.shape[0] < 3:
            break

    print(found_circle_centers)
    print(found_circle_radii)


    ############# Plot #############
    # x and y data
    x = data[:, 0]
    y = data[:, 1]

    # plot points with redouces marker size
    plt.scatter(x, y, s=1)
    plt.axis('square')

    for circle_center, circle_radius in zip(found_circle_centers, found_circle_radii):

        circle = plt.Circle((circle_center[0], circle_center[1]), radius=circle_radius, fill=False) # Create a circle
        plt.gca().add_patch(circle) # Add it to the plot

    title = r"$\epsilon$ = " + str(epsilon)
    plt.title(title)
    plt.show()
```
